```python
fig, axes = plt.subplots(2, 2, figsize=(15, 10))

sns.histplot(iris['sepal_length'], kde=True, ax=axes[0, 0])
axes[0, 0].set_title('Sepal Length Distribution')

sns.histplot(iris['sepal_width'], kde=True, ax=axes[0, 1])
axes[0, 1].set_title('Sepal Width Distribution')

sns.histplot(iris['petal_length'], kde=True, ax=axes[1, 0])
axes[1, 0].set_title('Petal Length Distribution')

sns.histplot(iris['petal_width'], kde=True, ax=axes[1, 1])
axes[1, 1].set_title('Petal Width Distribution')

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.boxplot(data=iris.drop('species', axis=1))
plt.title('Boxplot of Numerical Features')
plt.show()
```

```python
sns.countplot(x='species', data=iris)
plt.show()
```

```python
iris['species'].value_counts().plot.pie(
    explode=[0.1, 0.1, 0.1],
    autopct='%1.1f%%',
    shadow=True,
    figsize=(8,8)
)

plt.show()
```

```python
sns.boxplot(x='species', y='petal_length', data=iris)
plt.show()
```

```python
sns.boxplot(x='species', y='petal_length', data=iris)

sns.stripplot(
    x='species',
    y='petal_length',
    data=iris,
    jitter=True,
    edgecolor='gray'
)

plt.show()
```

```python
species = iris['species'].unique()
colors = ['red', 'green', 'blue']

plt.figure(figsize=(8,6))

for i in range(len(species)):
    x = iris[iris['species'] == species[i]]

    plt.scatter(
        x['sepal_length'],
        x['sepal_width'],
        c=colors[i],
        label=species[i]
    )

plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.title('Sepal Length vs Sepal Width')
plt.legend()
plt.show()
```

```python
sns.pairplot(iris, hue='species')
plt.show()
```

```python
corr = iris.drop('species', axis=1).corr()

plt.figure(figsize=(8,6))

sns.heatmap(
    corr,
    annot=True,
    cmap='coolwarm',
    fmt='.2f'
)

plt.title('Correlation Heatmap')
plt.show()
```
